```python
from selenium import webdriver  # 크롬 열고, 접속하고 이런 애들
from selenium.webdriver.common.by import By  # 태그 검색
from selenium.webdriver.common.keys import Keys  # 키보드 입력을 다루기 위해 필요
import time  # 시간 다루기 (3초 휴식)

browser = webdriver.Chrome() #크롬 열기

url = "https://www.naver.com/"

# 사이트 접속
browser.get(url)

#3초 쉰다
time.sleep(3)

# 자동스크롤
# "body"태그 찾아서 키보드 "PAGE_DOWN" 입
for i in range(3):
    browser.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
```

```python
# 필요한 모듈 불러오기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
import os
import urllib.request as req

# 검색할 키워드와 URL 설정
keyword = "곰인형"
url = f"https://www.google.co.kr/search?tbm=isch&q={keyword}"

# 크롬 웹드라이버 열기
browser = webdriver.Chrome()

# URL로 이동
browser.get(url)

# 스크롤을 5번 내리기 (스크롤 한번당 1초 대기)
for i in range(5):
    browser.find_element(By.TAG_NAME, "body").send_keys(Keys.PAGE_DOWN)
    time.sleep(1)

# 이미지 태그 찾기
imgs = browser.find_elements(By.CSS_SELECTOR, "div.H8Rx8c img")

# 이미지 다운로드 폴더 설정
folderName = f"{keyword} 검색결과"

# 폴더가 없으면 생성
if not os.path.exists(folderName):
    os.mkdir(folderName)

# 이미지 다운로드
num = 1
for img in imgs:
    imgUrl = img.get_attribute("src")  # 이미지 URL 가져오기
    path = f"{folderName}/{num}.jpg"  # 이미지 저장 경로 설정
    req.urlretrieve(imgUrl, path)  # 이미지 다운로드 및 저장
    num += 1  # 파일명 번호 증가
    print(path, "저장완료!")
```
