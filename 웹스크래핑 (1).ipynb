**크롤링 하는 순서**
1. request 모듈로 HTML코드 받아오기 -> requests로 import
2. BeautifulSoup 모듈로 HTML코드 구조화하기
3. CSS 선택자를 사용하여 원하는 태그 고르기
4. 반복이 필요하다면 규칙을 찾아서 for문 돌리기

```python
import requests
from bs4 import BeautifulSoup

# 웹 사이트 html 코드들 다 가져와서 code변수에 저장
code = requests.get("http://www.cgv.co.kr/movies/?lt=1&ft=0") #requests 주소 접속(요청)
# print(code) #<Response [200]>출력 -> HTTP의 상태코드를 보면ㄴㅇ 200번은 정상적으로 크롤링이 된 것을 알려주는 번호 (response로 응답(헤더 + 문서)이 온 것)

# print(code.headers) #헤더 정보(환경 정보)
# Python용 -> 딕셔너리로 불러옴 , web용 -> JSON
# print(code.text) #문서 정보(HTML)

# bs로 HTML코드 구조화하기, 주의! 우리 눈에는 티 안남 , 파이썬 내부적으로 구조화!
soup = BeautifulSoup(code.text, "html.parser")

# 영화제목은 strong태그 class속성이 title인거!
# css 선택자 id선택자 -> #속성명, class 선택자 .속성명
# 태그 찾기!
title = soup.select_one("strong.title")

# 태그 정보 출력
print(title)

# 태그 안에 글자만 출력
print(title.text)
```

```python
# 여러개 태그 찾기!
# class속성이 title인 strong태그 찾음!
title = soup.select("strong.title")

li_title = []

for title_name in title:
  print(title_name.text)
```

## 무신사 옷 평균 가격 구하기

```python
import requests
from bs4 import BeautifulSoup

# 웹 페이지 요청
code = requests.get("https://www.musinsa.com/brands/mahagrid?gclid=CjwKCAjwnqK1BhBvEiwAi7o0XyQfqWuLutLRXj6hItSt0gYqKERc7BCSp7t9xDCbDySNN7crJ3fjYBoCkCoQAvD_BwE")

# BeautifulSoup 객체 생성
soup = BeautifulSoup(code.text, 'html.parser')

# 필요한 정보 태그 찾기
title = soup.select("strong.brandshop-product__price__sale")
title2 = soup.select("a.brandshop-product__info__name.ampl-catch-click")

# 리스트로 저장(갯수 세기 위해서)
li_title = []
for price in title:
  li_title.append(price)

li_title2 = []
for name in title2:
  li_title2.append(name)

# 판매 상품 정보
li_imt = []
for name,price in zip(li_title2, li_title):
  # print(f"상품명:{name.text},가격{price.text}")
  li_imt.append(f"상품명:{name.text},가격{price.text}")

# 판매 상품 갯수 구하기
print(li_imt)
print(int(len(li_imt)) + 1)
```

```python
# 제품의 평균 가격 구하기
li_title = []
for prices in title:
  price = prices.text
  price1 = int(price.replace(",",""))
  li_title.append(price1)

average = sum(li_title) / len(li_title)+1
print(f"무신사 평균 옷 가격: {round(average,1)}원")
```

## 환율 가져오기

```python
import requests
from bs4 import BeautifulSoup

# <span class="value">900.61</span>

# request 모듈로 HTML코드 받아오기 -> requests로 import
code = requests.get("https://finance.naver.com/marketindex/")

# BeautifulSoup 모듈로 HTML코드 구조화하기
soup = BeautifulSoup(code.text, 'html.parser')

# CSS 선택자를 사용하여 원하는 태그 고르기
Exchange_Rate = soup.select("ul#exchangeList.data_lst span.value")
# print(Exchange_Rate)
# print(Exchange_Rate.text)


# 반복이 필요하다면 규칙을 찾아서 for문 돌리기
li_exchange_rate = []
for exchange_rate in Exchange_Rate:
  data = exchange_rate.text
  li_exchange_rate.append(data)
# print(li_exchange_rate)

# 보기 좋게 출력하기
print(f" 미국USD {li_exchange_rate[0]}\n 일본JPY {li_exchange_rate[1]}\n 유럽연합EUR {li_exchange_rate[2]}\n 중국CNY {li_exchange_rate[3]} ")
```

## 네이버 연관검색어 가져오기

```python
import requests

from bs4 import BeautifulSoup

keyword = input("검색어를 입력해주세요")

code = requests.get(f"https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=0&ie=utf8&query={keyword}")

soup = BeautifulSoup(code.text, "html.parser") #웹 사이트 소스 가져오기

price = soup.select("a.keyword div.tit") # html 구조화

for i in price:
    print(i.text)
```

## 지식IN 제목 수집

```python
import requests
from bs4 import BeautifulSoup
keyword = input()
code = requests.get(f"https://kin.naver.com/search/list.naver?query={keyword}")

soup = BeautifulSoup(code.text, "html.parser")

select_name = soup.select("dl dt")
select_data = soup.select("dd.txt_inline")

for title, data in zip(select_name,select_data): # for문에서 리스트 여러개로 반복 돌리기
  print(title.text, data.text)
```

#여러 페이지 수집

```python
import requests
from bs4 import BeautifulSoup
keyword = input()

for page in range(1, 11):
  code = requests.get(f"https://kin.naver.com/search/list.naver?query=파이썬잘하는법&page={page}")

  soup = BeautifulSoup(code.text, "html.parser")

  select_name = soup.select("dl dt")
  select_data = soup.select("dd.txt_inline")

  for title, data in zip(select_name,select_data): # for문에서 리스트 여러개로 반복 돌리기
    print(title.text, data.text)
```

```python
import requests
from bs4 import BeautifulSoup # 1. import

for i in range(1,11):
  page = requests.get(f"https://underkg.co.kr/index.php?mid=news&page={i}") # 2. 요청

  soup = BeautifulSoup(page.text, "html.parser") # 3. 응답

  title = soup.select("div.col-width.pull-left h1.title a") # 4. 선택자 선택해서 원하는 문구 가져오기1
  time = soup.select("span.time > span") # 4. 선택자 선택해서 원하는 문구 가져오기2

  for title_p, time_a in zip(title, time): # 5. for문 돌려서 1~11페이지 원하는 문구 가져오기
    print("제목: ",title_p.text)
    print("날짜 ", time_a.text)
    print("링크 ", title_p["href"]) #a태그 안에 있는 링크 가져오는 법
    link = title_p["href"]

    # 한번 더 링크 들어가서 수집
    page2 = requests.get(title_p["href"])

    soup = BeautifulSoup(page2.text, "html.parser")

    p_tag = soup.select_one("div.read_body")

    print(p_tag.text.replace("\n",""))

    print("=======================================================================================================================")
```
