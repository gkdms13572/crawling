```python
# 1. 관련 라이브러리 (모듈) 불러오기
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

# 2. Chrome 브라우저 열기
browser = webdriver.Chrome()

# 3. 크롤링하고 싶은 url 브라우저에 전송
url = 'https://www.melon.com/chart/index.htm'
browser.get(url)

time.sleep(2)
# 4. CSS 선택자로 원하는 태그 가져오기

# 차트 파인더 클릭
browser.find_element(By.CSS_SELECTOR,'button.btn_chart_f').click()
time.sleep(1)
# 주간차트 클릭
browser.find_element(By.CSS_SELECTOR,'h4.tab01').click()
time.sleep(1)
# 연대 클릭
browser.find_element(By.CSS_SELECTOR,'label[for=decade_2]').click()
time.sleep(1)
# 연도 클릭
browser.find_element(By.CSS_SELECTOR,'label[for=year_2]').click()
time.sleep(1)
# 월간 클릭
browser.find_element(By.CSS_SELECTOR,'label[for=month_2]').click()
time.sleep(1)
# 주간 클릭
browser.find_element(By.CSS_SELECTOR,'label[for=week_2]').click()
time.sleep(1)
# 장르 클릭
browser.find_element(By.CSS_SELECTOR,'label[for=gnr_1]').click()
time.sleep(1)
# 검색 클릭
browser.find_element(By.CSS_SELECTOR,'button.btn_b26 span.even_span').click()

time.sleep(1)

# 제목,가수이름 가져오기
title = browser.find_elements(By.CSS_SELECTOR,'div.rank01')
singers = browser.find_elements(By.CSS_SELECTOR,'div.rank02')

i = 0
for 제목, 가수 in zip(title, singers):
    i += 1
    print(f"순위 {i}위")
    print("제목 : ", 제목.text)
    print("가수 : ", 가수.text)
    print("===============================")
```
